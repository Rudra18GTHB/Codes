```python
import numpy as np
```

```python

class Perceptron:
    def __init__(self, inputs, output):
        self.inputs = inputs
        self.weight_matrix = np.array([0.7,0.2,0.1])
        self.bias = 5
        self.output = output

    def weighted_sum(self):
        sum = np.matmul(self.inputs, self.weight_matrix.T) + self.bias
        return sum
    
    def activation_func(self,sum ):
        if sum>0:
            return 1
        if sum <0:
            return 0
    
    def predict(self):
        sum = self.weighted_sum()
        self.predicted = []
        for i in range(sum.shape[0]):
            self.predicted.append(self.activation_func(sum[i]))
        return self.predicted
        
    def calculate_loss(self):
        self.loss = np.sum(np.array(self.predict()) - np.array(self.output))
        return self.loss

    def weight_updates(self):
        self.weight_matrix = self.weight_matrix + 0.1*self.calculate_loss()
        self.bias = self.bias + 0.1*self.calculate_loss()

    def fit(self):
        self.num_epoch = 100
        for i in range(self.num_epoch):
            self.weight_updates()
            print(f"Loss after epoch {i} is {self.loss}")
```

```python
input_data = np.array([[10,3,3],
                        [12,4,4],
                        [15,2,2],
                        [20,1,1],
                        [10,5,5],
                        [8,5,5]])
output = [1,1,0,0,1,1]
```

```python

perceptron1 = Perceptron(input_data, output)
```

```python
perceptron1.fit()
```
