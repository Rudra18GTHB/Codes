```python
import cv2
import time
```

```python
image_folder = "test\\Left Hand\\image"
# Initialize webcam
cap = cv2.VideoCapture(0)
photo_count = 1
for i in range(100):
    ret, frame = cap.read()  # Read frame from webcam
    # Resize the frame
    resized_frame = cv2.resize(frame, (224, 224))
    image_path = image_folder + str(photo_count) +".jpg"
    cv2.imwrite(image_path, resized_frame)
    photo_count+=1
    print(f"Photo saved:{photo_count}")
    time.sleep(0.2)
cap.release()
```
