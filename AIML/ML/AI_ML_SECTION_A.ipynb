```python
graph ={
    '5' :  ['3','7'],
    '3' : ['2','4'],
    '7' : ['8'],
    '2' : [],
    '4' : ['8'],
    '8' : []
}
print(graph)
```

```python
# prompt: problem solving by searching algorith bfs and dfs in ai
# BFS algorithm
def bfs(graph, start):
  visited = []
  queue = [start]
  while queue:
    node = queue.pop(0)
    if node not in visited:
      visited.append(node)
      for neighbor in graph[node]:
        queue.append(neighbor)
  return visited

# DFS algorithm
def dfs(graph, start):
  visited = []
  stack = [start]
  while stack:
    node = stack.pop()
    if node not in visited:
      visited.append(node)
      for neighbor in graph[node]:
        stack.append(neighbor)
  return visited

# Print the BFS and DFS traversal of the given graph
print("BFS traversal:", bfs(graph, '5'))
print("DFS traversal:", dfs(graph, '5'))
```

#**EXPERIMENT 10**

implementation of k means clustring

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
u=[4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
v = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
data = list(zip(u,v))
print(data)
```

```python
# order to find the best value for K, we need to run K-means across our data for a range of possible values. We only have 10 data points, so the maximum number of clusters is 10. So for each value K in range(1,11), we train a K-means model and plot the intertia at that number of clusters:
inertias = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(data)
  inertias.append(kmeans.inertia_)
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# prompt: perform hill climbing search of a one-dimensional objective function. in numpy

import numpy as np

def hill_climbing(objective_function, initial_state, max_iterations=1000, tolerance=1e-6):
    """
    Performs hill climbing search on a one-dimensional objective function.

    Args:
        objective_function: The objective function to be optimized.
        initial_state: The initial state of the search.
        max_iterations: The maximum number of iterations to perform.
        tolerance: The tolerance for convergence.

    Returns:
        The best state found, the best objective value, and the number of iterations performed.
    """

    # Initialize the current state and value
    current_state = initial_state
    current_value = objective_function(current_state)

    # Iterate until convergence or maximum iterations reached
    iterations = 0
    while True:
        # Generate neighboring states
        neighbors = [current_state + 1, current_state - 1]

        # Find the best neighbor
        best_neighbor = None
        best_value = None
        for neighbor in neighbors:
            neighbor_value = objective_function(neighbor)
            if best_neighbor is None or neighbor_value > best_value:
                best_neighbor = neighbor
                best_value = neighbor_value

        # Check for convergence
        if abs(best_value - current_value) < tolerance:
            break

        # Update the current state and value
        current_state = best_neighbor
        current_value = best_value

        iterations += 1

        # Check for maximum iterations
        if iterations >= max_iterations:
            break

    return current_state, current_value, iterations

# Define the objective function
def objective_function(x):
    return -np.sin(x)

# Perform hill climbing search
best_state, best_value, iterations = hill_climbing(objective_function, 0)

# Print the results
print("Best state:", best_state)
print("Best value:", best_value)
print("Iterations:", iterations)
```

```python
def objective_function(x):
    # Define your objective function here
    return -(x**2)  # Example: maximizing a quadratic function

def hill_climbing(initial_position, step_size, max_iterations):
    current_position = initial_position

    for iteration in range(max_iterations):
        current_value = objective_function(current_position)

        # Move to the next position by taking a step in the direction of increasing values
        next_position = current_position + step_size

        # Check if moving to the next position improves the objective function value
        next_value = objective_function(next_position)
        if next_value > current_value:
            current_position = next_position
        else:
            # Break the loop if no improvement is found
            break

    return current_position, objective_function(current_position)

# Example usage
initial_position = 2.0
step_size = 0.1
max_iterations = 100

result_position, result_value = hill_climbing(initial_position, step_size, max_iterations)

print(f"Optimal Position: {result_position}")
print(f"Optimal Value: {result_value}")
```

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
data = load_iris()
X, y, column_names = data['data'], data['target'], data['feature_names']
X = pd.DataFrame (X, columns = column_names)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y, random_state = 44)
```

```python
means = X_train.groupby (y_train).apply(np.mean)
stds = X_train.groupby(y_train).apply(np.std)
probs = X_train.groupby(y_train).apply(lambda x: len(x)) /X_train.shape[0]
```

```python
import scipy.stats as norm
y_pred = []
for elem in range(X_val.shape[0]):
  p = {}
  for cl in np.unique(y_train):
    p[cl] = probs.iloc[cl]
    for index, param in enumerate(X_val.iloc[elem]):
      p[cl] *= norm.pdf(param, means.iloc[cl, index], stds.iloc[cl, index])
  y_pred.append(pd.Series(p).values.argmax())
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_val, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
accuracy_score(y_val, model.predict(X_val))
```
