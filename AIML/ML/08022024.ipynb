```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("DataSet/IPL IMB381IPL2013.csv")
df
```

```python
df.columns
```

```python
X=df.columns
X=['AGE', 'COUNTRY', 'PLAYING ROLE','T-RUNS', 'T-WKTS', 'ODI-RUNS-S', 'ODI-SR-B', 'ODI-WKTS', 'ODI-SR-BL', 'CAPTAINCY EXP', 'RUNS-S', 'HS', 'AVE', 'SR-B', 'SIXERS', 'RUNS-C', 'WKTS', 'AVE-BL', 'ECON', 'SR-BL']
```

```python
categorical = ['AGE', 'COUNTRY','PLAYING ROLE', 'CAPTAINCY EXP',]
```

```python
encoded_df = pd.get_dummies(df[X], columns = categorical, drop_first = True)
```

```python
encoded_df.columns
```

```python
X = encoded_df.columns
```

```python
x = sm.add_constant(encoded_df)
y = df['SOLD PRICE']

train_X,test_X,train_Y, test_Y = train_test_split(x,y, train_size =0.8, random_state = 42)
```

```python
model = sm.OLS(train_Y, train_X).fit()
model.summary2()
```
