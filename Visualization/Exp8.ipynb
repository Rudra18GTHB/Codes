```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("batter.csv")
df.head(25)
```

```python
fig, ax=plt.subplots(nrows=2,ncols=1, figsize=(8,8))
ax[0].scatter(df['avg'],df['strike_rate'],color='red')
ax[1].scatter(df['avg'],df['runs'],color='yellow')
ax[0].set_title("IPL Analysis")
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')
ax[1].set_xlabel('Runs')
ax[1].set_ylabel('Strike Rate')
fig.show()
```
