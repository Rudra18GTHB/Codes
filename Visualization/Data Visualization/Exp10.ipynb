```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.array([12,45,78,23])
y=np.arange(0,20,5)
z=np.linspace(0,20,4,dtype=int)
```

```python
fig=plt.figure(figsize=(12,8),facecolor="pink")
ax=plt.axes(projection="3d")
ax.set_facecolor("Gray")
ax.scatter3D(x,y,z, color="black", marker="d",depthshade=False)
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_zlabel("z axis")
#ax.view_init(45,0)
ax.set_title("Scatter Plot")
ax.plot3D(x,y,z, color="cyan")
fig.show()
```
