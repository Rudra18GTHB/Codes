```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns;
```

```python
plt.boxplot([10,20,30,40,50,60,70,120],vert=True,notch=True,showmeans=True,patch_artist=True,boxprops={'color':"red"},capprops={'color':'Orange'},whiskerprops={'color':'Yellow'},sym="g+")
plt.show()
```

```python
df=sns.load_dataset("iris")
df
```

```python
plt.boxplot([df["sepal_length"],df["sepal_width"],df["petal_length"],df["petal_width"]],vert=False,showmeans=True,notch=True,patch_artist=True,sym="r*",boxprops={'color':"pink","facecolor":"lightblue"},capprops={'color':'Orange'},whiskerprops={'color':'Orange'},labels=["Sepal Length","Sepal Width","Petal Length","Petal Width"])
plt.title("Species")
plt.show()
```

```python
df=sns.load_dataset("tips")
df
```

```python
plt.boxplot([df["total_bill"],df["tip"]],sym="y*",showmeans=True,patch_artist=True,boxprops={'color':"pink","facecolor":"lightblue"},capprops={'color':'Orange'},whiskerprops={'color':'Orange'},labels=["Bill","Tip"])
plt.title("Tips")
plt.show()
```
