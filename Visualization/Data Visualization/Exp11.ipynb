```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([12,45,78,23,15,20,73,99])
y=np.arange(0,24,3)
x, y = np.meshgrid(x,y)
z=np.sqrt(x**2+y**2)
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
p=ax.plot_surface(x,y,z, cmap='plasma')
plt.colorbar
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.subplot()
ax.contour(x,y,z,cmap="plasma")
plt.show()
```
