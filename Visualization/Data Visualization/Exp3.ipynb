```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# make data set using list
product1=np.random.randint(0,100,23)
year1=np.linspace(2000,2023,23)
```

```python
product2=np.random.randint(0,100,23)
year2=np.linspace(2000,2023,23)
```

```python
plt.scatter(year1,product1,label="product1")
plt.scatter(year2,product2,label="product2")
plt.title("Product Sales")
plt.xlabel("Year")
plt.ylabel("Product")
plt.legend()
plt.show()
```

```python
df=pd.read_csv("batter.csv")
df
```
