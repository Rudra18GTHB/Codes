```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([23,45,6,7,8,91,9,78,68,5,49,3,45])
y = np.array([34,56,23,78,43,23,88,11,20,55,67,3,6])
print(x)
print(y)
```

```python
fig, ax= plt.subplots(nrows=2,ncols=2,figsize=(15,7))
ax[0,0].scatter(x,y,color="blue")
ax[0,0].set_title("Scatter Plot")
ax[0,0].set_xlabel("x")
ax[0,0].set_ylabel("y")
ax[0,1].bar(x,y,color="red")
ax[0,1].set_title("Bar Plot")
ax[0,1].set_xlabel("x")
ax[0,1].set_ylabel("y")
ax[1,0].pie(x,labels=y)
ax[1,0].set_title("Pie Plot")
ax[1,1].hist(x,bins=[10,20,30,40,50,60,70,80,90],color="pink")
ax[1,1].set_title("Histogram")
ax[1,1].set_xlabel("x")
ax[1,1].set_ylabel("y")
fig.show()
```
