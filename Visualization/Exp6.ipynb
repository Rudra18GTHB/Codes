```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use("classic")
```

```python
#Univariate /bivariate analysis
#categorical and numerical accepted
# To find contribution on a standard basis
```

```python
marks=[86,79,45,89,99,63]
subjects=["Maths","Computer","Chemistry","Physics","English","Social"]
plt.pie(marks,labels=subjects,autopct="%0.2f%%",colors=["green","red","blue","purple","orange","yellow"],startangle=90,explode=[0.05,0,0,0,0,0],shadow=True)
plt.legend(loc="upper left",fontsize=5)
plt.show()
```

```python
data_scientists_salary=[6500,9280,12050,13200,16672,21000,23965,29793]
data_scientists_name=["John","Peter","Tony","Banner","Steve","May","Nick","Jarvis"]
```

```python
plt.pie(data_scientists_salary,
        labels=data_scientists_name,
        autopct="%0.2f%%",
        colors=["green","red","blue","purple","orange","yellow","lightgreen","lightblue"],
        startangle=90,
        shadow=True,
        )
plt.title("Salary Analysis")
plt.show()
```
