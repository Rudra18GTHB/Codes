```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
product=[40111,45012,45625,40254,41254]
year=[2019,2020,2021,2022,2023]
```

```python
plt.plot(year,product,color="#ed0933",linewidth=2,marker="s",markersize=6,label="sales")
plt.title("Sales")
plt.xlabel("Year")
plt.ylabel("Products")
plt.legend(loc="best")
plt.show()
```

```python
df=pd.read_csv("sharma-kohli.csv")
df
```

```python
plt.plot(df["index"],df["RG Sharma"],label="Rohit",color="green",linewidth="2",linestyle="solid",marker="o")
plt.plot(df["index"],df["V Kohli"],label="Virat",color="red",linewidth="2",linestyle="solid",marker="o")
plt.title("IPL Analysis")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
plt.show()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv")
df
```

```python
plt.plot(df["Year"],df["USA"],label="AUS",color="green",linewidth="2",linestyle="solid",marker="o")
plt.plot(df["Year"],df["Canada"],label="CAN",color="red",linewidth="2",linestyle="solid",marker="o")
plt.title("Gas Price Analysis")
plt.xlabel("Year")
plt.ylabel("Gas Price")
plt.legend()
plt.show()
```
