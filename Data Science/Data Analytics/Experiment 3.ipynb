```python
import matplotlib.pyplot as plt
```

```python
# Data to plot
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
sizes = [35, 25, 20, 15, 5]  # Percentages
colors = ['blue', 'red', 'green', 'yellow', 'orange']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

```

```python
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Data for the pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create the figure and axis
fig, ax = plt.subplots()

# Function to update the pie chart for animation
def update(num):
    ax.clear()  # Clear the previous pie chart
    sizes_updated = [size + num for size in sizes]  # Modify the sizes to animate
    ax.pie(sizes_updated, labels=labels, autopct='%1.1f%%', startangle=90)
    ax.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.

# Create the animation
ani = FuncAnimation(fig, update, frames=range(40), interval=20)

# Display the animation in Google Colab
HTML(ani.to_html5_video())
```
