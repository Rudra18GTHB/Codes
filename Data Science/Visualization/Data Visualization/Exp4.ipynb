```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#problem statement
experience=[1,3,4,5,7,9,10,12]
TCS=[6900,9200,12050,13298,16672,21000,23965,29793]
Wipro=[5500,9388,12450,14290,13472,22998,25965,33793]
Info=[6700,8322,15000,13900,19039,26998,30965,43793]
```

```python
#tcs
plt.bar(experience,TCS,color="m",width=0.8,edgecolor="blue")
plt.xlabel("experience(in years)")
plt.ylabel("salary")
plt.title("TCS salary chart")
plt.grid()
plt.show()
```

```python
#wipro
plt.bar(experience,Wipro,color="yellow",width=0.8,edgecolor="red")
plt.xlabel("experience(in years)")
plt.ylabel("salary")
plt.title("Wipro salary chart")
plt.grid()
plt.show()
```

```python
plt.barh(experience,Info,color="green",edgecolor="black")
plt.ylabel("experience(in years)")
plt.xlabel("salary")
plt.title("Info salary chart")
plt.grid()
plt.show()
```

```python
#grouped chart
#plt.bar(experience,TCS,color="red")
w=0.2
x=np.arange(len(experience))
x
x1=[i+w for i in x]

x2=[i-w for i in x]

plt.bar(x,TCS,color="red",width=w,edgecolor="yellow",label="TCS")
plt.bar(x1,Wipro,color="pink",width=w,edgecolor="black",label="Wipro")
plt.bar(x2,Info,color="green",width=w,edgecolor="blue",label="Info")
plt.ylabel("salary")
plt.xlabel("experience in years")
plt.title("Salary of TCS,Wipro,Info")
plt.legend()
plt.grid()
plt.show()

```
