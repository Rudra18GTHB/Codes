```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("iris (2).csv")
df.sample(5)
```

```python
df["Species"]=df["Species"].replace({"Iris-setosa":0,"Iris-virginica":1,"Iris-versicolor":2})
df.sample(5)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap="jet")
plt.title("Species")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Width")
plt.colorbar()
plt.show()
```

```python
x=[1,5,3,8,9]
y=[1,5,3,8,9]
plt.scatter(x,y)
plt.text(1,1,"Point1")
plt.show()
```

```python
df=pd.read_csv("batter.csv")
df=df.head(25)
df
```

```python
plt.scatter(df["avg"],df["strike_rate"],s=df['runs']/20)
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
plt.title("IPL Analysis")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.show()
```
